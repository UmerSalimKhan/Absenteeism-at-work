### Grouping

<h4>In this section, we will:</h4>
<ul>
    <li>Group various reasons into fewer groups.</li>
</ul>

Reason for absence (ICD).
	Absences attested by the International Code of Diseases (ICD) stratified into 21 categories (I to XXI) as follows:
	I Certain infectious and parasitic diseases
	II Neoplasms
	III Diseases of the blood and blood-forming organs and certain disorders involving the immune mechanism
	IV Endocrine, nutritional and metabolic diseases
	V Mental and behavioural disorders
	VI Diseases of the nervous system
	VII Diseases of the eye and adnexa
	VIII Diseases of the ear and mastoid process
	IX Diseases of the circulatory system
	X Diseases of the respiratory system
	XI Diseases of the digestive system
	XII Diseases of the skin and subcutaneous tissue
	XIII Diseases of the musculoskeletal system and connective tissue
	XIV Diseases of the genitourinary system
	XV Pregnancy, childbirth and the puerperium
	XVI Certain conditions originating in the perinatal period
	XVII Congenital malformations, deformations and chromosomal abnormalities
	XVIII Symptoms, signs and abnormal clinical and laboratory findings, not elsewhere classified
	XIX Injury, poisoning and certain other consequences of external causes
	XX External causes of morbidity and mortality
	XXI Factors influencing health status and contact with health services.

	And 7 categories without (CID) patient follow-up (22), medical consultation (23), blood donation (24), laboratory examination (25), unjustified absence (26), 	physiotherapy (27), dental consultation (28).


<h3>Groups:</h3>
> 1 - 5 : Serious diseases (0), <br>
> 6 - 14 : Diseases (1), <br>
> 15 - 17 : Pregnancy (2), <br>
> 18 - 21 : Health issues (3), <br> 
> 22 - 28 : Extra issues (4) <br>

````python
# Reason for absence
def group_reasons(df, column_to_be_treated):
    """
    Groups the 'Reason for absence' column into specified categories.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        column_to_be_treated (str): The column name to group.

    Returns:
        pd.DataFrame: The modified DataFrame with grouped reasons.
    """
    # Define the grouping logic
    def categorize_reason(reason):
        if 1 <= reason <= 5:
            return 0  # Serious diseases
        elif 6 <= reason <= 14:
            return 1  # Diseases
        elif 15 <= reason <= 17:
            return 2  # Pregnancy
        elif 18 <= reason <= 21:
            return 3  # Health issues
        elif 22 <= reason <= 28:
            return 4  # Extra issues
        else:
            return None  # Handle invalid/missing values

    # Apply the grouping function to the column
    df[column_to_be_treated] = df[column_to_be_treated].apply(categorize_reason)
    return df
    
````

````python
from sklearn.preprocessing import StandardScaler

# Select columns to standardize
columns_to_standardize = [col for col in df_encoded.columns if col not in one_hot_columns]

# Initialize scaler
scaler = StandardScaler()

# Standardize only the selected columns
df_encoded[columns_to_standardize] = scaler.fit_transform(df_encoded[columns_to_standardize])

# Checkpoint 
df_scaled = df_encoded.copy()

# Output standardized DataFrame
display(df_encoded.head())
````

# Grouping target column

````python
df_scaled_encoded = df_scaled.copy()
````

````python
# If Absenteeism time in hours is less than to one hours than, we will consider it to be a non-significant gap
# Else not

# Non-significant gap = 0
# Significant = 1

def grouping_absenteeism_time_in_hours(val):
    if val <= 1:
        return 0
    elif val > 1:
        return 1

df_scaled_encoded['Absenteeism time in hours'] = df_scaled_encoded['Absenteeism time in hours'].apply(grouping_absenteeism_time_in_hours)
````
